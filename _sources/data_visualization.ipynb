```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-3, 3, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.show()

plt.plot(x, y2)
plt.show()
```
